```python
import numpy as np
import matplotlib.pyplot as plt
```

## signal
this notebooks checks the content of the collected data files.

```python
def get_data(zbin):
    r, wp = np.loadtxt('../data/wp_sig_z{:03d}.dat'.format(zbin), unpack=True)
    cov = np.loadtxt('../data/wp_cov_z{:03d}_z{:03d}.dat'.format(zbin, zbin))
    return r, wp, cov
```

```python
for zbin in range(3):
    r, wp, cov = get_data(zbin)
    std = np.sqrt(np.diag(cov))
    plt.errorbar(r*1.02**zbin, r*wp, r*std, fmt='.')
plt.xscale('log')
```
